```python
class TicTacToe:
    
    def __init__(self,game_size, play = True, choice = 1, game_won = False, played = False):
        self.game_size = game_size
        self.game = [" " for x in range(self.game_size**2)]
        self.play = play
        self.choice = choice
        self.game_won = game_won
        self.played = played
    
    def display(self):
        count = 1
        print("---------------")
        for i in range(self.game_size**2):
            print("|",format(self.game[i], "2"), end=" ")
            if count % self.game_size == 0:
                print()
                print("---------------")
            count +=1
    
    def game_board(self,player = 0, pos = 0):
        try:
            if self.game [pos] != " ":
                print("This position is Occupied! Choose another!")
                return False
            else:
                self.game [pos] = player
            return True

        except IndexError as e:
            print("Make sure your input row/column is 0 1 or 2 ", e)
            return False

        except Exception as e:
            print("Something is not right", e)
            return False
    
    def is_draw(self):
        used_pos = 0
        for loc in self.game:
            if loc != " ":
                used_pos += 1
        if used_pos == self.game_size**2:
            return True
    
    def is_win(self):
        
        def all_same(lst):
            if lst.count(lst[0]) == len(lst) and lst[0] != " ":
                return True
            else:
                return False
        
        # Horizontal Winner 
        for i in range(0,self.game_size**2,3):
            if (self.game [i] == self.game [i+1] == self.game [i+2]) and self.game[i] !=" " :
                print (" Player", self.game[i], "is the Winner Horizontally!")
                return True
        
        # Vertical Winner 
        for i in range(0,self.game_size):
            if (self.game [i] == self.game [i+3] == self.game [i+6]) and self.game[i] !=" " :
                print (" Player", self.game[i], "is the Winner Vertically!")
                return True
        
        #Diagnoal Winner Left to Right bottom
        diags1 = []
        for i in range (0,self.game_size**2,self.game_size+1):
            diags1.append(self.game[i])
        if all_same(diags1):
            print("Player",diags1[0]," is the Winner Diagnoally! ")
            return True
        
        #Diagnoal Winner Right to Left bottom
        diags2 = []
        for i in range (self.game_size-1,self.game_size**2-1,self.game_size-1):
            diags2.append(self.game[i])
        if all_same(diags2):
            print("Player",diags2[0]," is the Winner Diagnoally! ")
            return True
        
        return False
     
    
    def reset(self):
        self.game = [" " for x in range(self.game_size**2)]

    def print_header(self):
        print("Welcome to Tic-Tac-Toe!\n")

def main():
    tic = TicTacToe(int(input("Enter the size of Tic-Tac-Toe game you want to play!: ")))
    players = [1,0]
    tic.print_header()
    tic.display() 

    while not tic.game_won and tic.play:
        current_player = tic.choice
        print("Current Player is: " , current_player)
        tic.choice = players[tic.choice]
        tic.played = False

        while not tic.played:  
            pos_choice = int(input("What position do you want to play ? "))
            tic.played = tic.game_board(current_player,pos_choice)  
            tic.display()

        if tic.is_win():
            print("Game Over!")
            tic.game_won = True
            tic.play = False
            tic.reset()

        if tic.is_draw():
            print("It's a Tie Game")
            tic.game_won = True
            tic.play = False
            tic.reset()
                
main()
```
